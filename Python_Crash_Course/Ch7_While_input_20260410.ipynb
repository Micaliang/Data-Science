# List to List moving

```python
# 薄弱题
unconfirmed_users = ['alice', 'brian', 'candace']
confirmed_users = []

while unconfirmed_users:
    current_users=unconfirmed_users.pop()

    print("Verifying users:" + current_users.title())
    confirmed_users.append(current_users)

print("\nThe following users have been confirmed:")
for confirmed_user in confirmed_users:
    print(confirmed_user.title())
```

```python
# 练习题
pending_orders = ['black tea','orange juice','milk tee']
complete_orders = []

print("These orders have been confirmed:")
while pending_orders:
    current_orders = pending_orders.pop()
    print("-Confirmed Order:" + current_orders.title())
    complete_orders.append(current_orders)

print("\nThese orders have been completed:")
for complete_order in complete_orders:
    print("-" + complete_order.title())
```

# Remove specific value from a list

```python
# 薄弱题
pets = ['dog', 'cat', 'dog', 'goldfish', 'cat', 'rabbit', 'cat']
print(pets)

while 'cat' in pets:  # 注意cat加符号‘’，用while不是for，因为while可以判断，for用来循环
    pets.remove('cat')
print(pets)
```

```python
# while + list + remove
fruits=['apple','orange','mango','banana','orange','watermelon','orange']
while 'orange' in fruits:
    fruits.remove('orange')
print(fruits)
```

# Filling a Dictionary with User Input

```python
# 薄弱题
responses={}
polling_active=True

while polling_active:
    name=input("\nWhat is your name?")
    response=input("Which mountain would you like to climb someday?")
    responses[name]=response

    repeat=input("Would you like to let another person respond? (yes/ no) ")
    if repeat=='no':
        polling_active=False

print("\n--- Poll Results ---")
for name,response in responses.items():
    print(name+" would you like to climb "+response+".")
```

```python
# while + dictionary + flag + input 
colors = {}
flag = True

while flag:
    name = input("What's your name?")
    response = input("What's your favorite color?")
    colors[name] = response

    repeat = input("Do you want someone to answer?")
    if repeat.lower() == 'no':
        flag = False

print("\n--- The result as bellow: ---")
for k,v in colors.items():
    print(k.title() + "'s favorite color is " + v + ".")
```

```python
# while + list + flag + input 
# 7-8 薄弱题
sandwich_orders=['tomato','beef','vegetable','lamb']
finished_sandwiches=[]

while sandwich_orders:
    current_sandwich=sandwich_orders.pop()
    print("I made your "+ current_sandwich+" sandwich")
    finished_sandwiches.append(current_sandwich)

print("\n---Sandwich that was made:---")
for finished_sandwich in finished_sandwiches:
    print(finished_sandwich + " sandwich was made")
```
