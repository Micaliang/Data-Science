# Creating Series

# From list

```python
list=[10,20,30,40,50]
label=['a','b','c','d','e']
s1=pd.Series(list,index=label)
print(s1)
```

# From 1D

```python
# 字典必须在创建的时候就定义key，不能像list和1维数组那样用label定义
narr=np.array([10,20,30,40,50])
s2=pd.Series(narr,index=label)
print(s2)
```

# From dictionary

```python
dict={'a':10,'b':20,'c':30,'d':40,'e':50}
s3=pd.Series(dict)
print(s3)
```

```python
s3.dtypes
```

```python
s3.info
```

```python
s3.size
```

```python
#转为float
s3.astype(float)
```

```python
#字典的drop需要删除的是key，其他直接drop索引
s3.drop('b')
```

# Creating DataFrame

# From list of lists

```python
# 注意columns需要加是复数s！！！
list=[[1,2],[3,4],[5,6]]
df1=pd.DataFrame(list,index=['a','b','c'],columns=['col_1','col_2'])
print(df1)
```

# From 2D ndarray

```python
ndarr=np.array([[1,2],[3,4],[5,6]])
df2=pd.DataFrame(ndarr,index=['a','b','c'],columns=['col_1','col_2'])
print(df2)
```

# From dictionary

```python
df3=pd.DataFrame({'R1':[1,2,3],'R2':[4,5,6]},index=['a','b','c'])
print(df3)
```

# Indexing Data in a Series

```python
dict={'a':10,'b':20,'c':30,'d':40,'e':50}
s3=pd.Series(dict)
print(s3)
```

```python
s3['b']
```

```python
s3.iloc[1]
```

```python
s3[['a','c']]
```

```python
# iloc() 是索引位置
s3.iloc[[1,2]]
```

```python
s3[s3>20]
```

# Indexing in a DataFrame

```python
list=[[1,2],[3,4],[5,6]]
df1=pd.DataFrame(list,index=['a','b','c'],columns=['col_1','col_2'])
print(df1)
```

```python
# 一个中括号
df1['col_1']
```

```python
# 两个中括号
df1[['col_1','col_2']]
```

# loc

```python
df1.loc['a']
```

# iloc

```python
df1.iloc[2]
```

```python
df1.iloc[1,1]
```
