# 1. Positional Arguments

# 2. Keyword Arguments

```python
def describe_pet(animal_type,pet_name):
    print("\nI have a "+ animal_type + ".")
    print("My " + animal_type + "'s name is " + pet_name.title() + ".")
 
describe_pet(animal_type='hamster', pet_name='harry')
describe_pet('monkey', pet_name='king')
describe_pet(pet_name='lily',animal_type='dog')
describe_pet('bird', 'sunny')

# 下面这两个会报错，因为不能指定最后一个参数
# describe_pet(pet_name='lily','dog') 违反关键词规则，关键词必须放在后面
# describe_pet('lucy',animal_type='bird') 指定了type，但是按顺序lucy有给type了，故报错
```

# Making an Argument Optional

```python
# 注意赋值的一定要放在最后
def get_formatted_name(first_name,last_name,middle_name=''):
    if middle_name:
        full_name=first_name + " " + middle_name + " " +last_name
    else:
        full_name=first_name + " " +last_name
    return full_name.title()
    
musician=get_formatted_name("mica","liang")
print(musician)

musician = get_formatted_name('john', 'hooker', 'lee')
print(musician)
```

# def + list 逆序输出 + 描述语(改变原列表)

```python
# 8-10 解法2 薄弱题
def show_magicians(names):
    for name in names:
        print(name.title())

def make_great(names):
    great_magicians = []
    
    while names:  # names非空时执行
        current_magician = names.pop()     # jessica-anna-mica
        great_magician = "the Great " + current_magician  # the Great + (jessica-anna-mica)
        great_magicians.append(great_magician) # the Great + (jessica-anna-mica)
    
    # 把加好问候语的名字放回列表
    for great_magician in great_magicians:
        names.append(great_magician)

magicians = ['mica','anna','jessica']

make_great(magicians) # 先加问候语
show_magicians(magicians) # 再输出
```

# def + list 顺序输出 + 描述语(改变原列表)

```python
# 8-10 顺序不变 薄弱题
def show_magicians(names):
    for name in names:
        print(name.title())

def make_great(names):
    great_magicians = []
    
    while names:
        current_magician = names.pop()
        great_magician = "the Great " + current_magician
        great_magicians.append(great_magician)
    
    # 倒序放回去，保持原来顺序
    while great_magicians:
        names.append(great_magicians.pop())

magicians = ['mica','anna','jessica']

make_great(magicians)
show_magicians(magicians)
```

# def + list copy输出 + 描述语(不改变原列表)

```python
# 薄弱题
def show_magicians(names):
    for name in names:
        print(name.title())

def make_great(names):
    # 建立一个空列表，存放加了称号的魔术师
    great_magicians = []

    # 把每个名字取出来，加上 "the Great"
    while names:
        current = names.pop()
        great_name = "the Great " + current
        great_magicians.append(great_name)
    
    # 返回新列表
    return great_magicians


# 原始列表
magicians = ['mica', 'anna', 'jessica']

# 重点：传副本！不修改原列表
great_magicians = make_great(magicians[:])

# 显示原始列表（没变）
print("Original magicians:")
show_magicians(magicians)

# 显示新列表（加了称号）
print("\nGreat magicians:")
show_magicians(great_magicians)
```

# def + **字典参数(两颗星)

```python
def build_profile(first,last,**user_info):
    profile={}
    profile['first_name']=first
    profile['last_name']=last
    for key,value in user_info.items():
        profile[key]=value
    return profile

user_profile=build_profile('albert', 'einstein',location='princeton',field='physics')
print(user_profile)
```

```python
def visit(first,last,**visitor_infos):
    visit_cloud={}
    visit_cloud['first_name']=first
    visit_cloud['last_name']=last

    for key,value in visitor_infos.items():
        visit_cloud[key]=value
    return visit_cloud

visitor=visit('mica','liang',country='China',city='Beijing')
print(visitor)
```
