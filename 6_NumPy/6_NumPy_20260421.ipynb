# Array Manipulations

```python
# reshape()
a=np.array([1,2,3,4,5,6])
a=a.reshape(2,3)
print(a)
```

```python
# concatenate()  一定要2个括号
a1=np.array([1,2,3])
a2=np.array([2,3,4])
np.concatenate((a1,a2),axis=0)
```

```python
# 垂直
a3=np.array([[1,2],[3,4]])
a4=np.array([[5,6],[7,8]])
np.concatenate((a3,a4),axis=0)
```

```python
# 水平
np.concatenate((a3,a4),axis=1)
```

```python
# stack
# 垂直
a3=np.array([[1,2],[3,4]])
a4=np.array([[5,6],[7,8]])
np.vstack((a3,a4))
```

```python
# 水平
np.hstack((a3,a4))
```

```python
# split
a=np.array([1,2,3,4,5,6])
np.split(a,3,axis=0)
```

```python
a3=np.array([[1,2],[3,4]])
np.split(a3,2,axis=0)
```

# View and Copy

# View

```python
arr=np.arange(5)
print(arr)
view_arr=arr[1:4] #basic slicing
print(view_arr)
view_arr[0]=999
print("Array after tempering:")
print("view array")
print(arr)
print("original array")
print(view_arr)
```

# copy

```python
arr=np.arange(5)
print(arr)
copy_arr=arr[1:4].copy()
print(view_arr)
copy_arr[0]=999
print("Array after tempering:")
print("original array")
print(arr)
print("copy array")
print(copy_arr)
```

```python
arr=np.arange(5)
print(arr)
view_arr=arr[[2,3,4]] # fancy indexing
#view_arr=arr[[2,3,4]].copy()
print(view_arr)
view_arr[0]=999
print("Array after tempering:")
print("view array")
print(arr)
print("original array")
print(view_arr)
```

# Excercise

```python
grades=np.array([85,92,78,95,88,76,91,83])
print("Grades",grades)
```

```python
# who got A(>=80)
mark_a=np.where(grades>=80)
print(grades[mark_a])
print(mark_a)
```

```python
# number of students who got grade>=90
mark_best=np.where(grades>=90)
print(grades[mark_best])
np.sum(grades>=90)
```

```python
# number of students who got grade>=90
mark_best=grades[(grades>=90)]
mark_best.size
```

```python
# find the student with top mark
np.argmax(grades)
```
