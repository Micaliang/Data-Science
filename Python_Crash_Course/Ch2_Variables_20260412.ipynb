# 运算符名称

```python
print(3 + 2)    # add 加法
print(5 - 3)    # subtract 减法
print(4 * 2)    # multiply 乘法
print(7 / 2)    # divide 除法
print(7 // 2)   # floor divide 整除
print(7 % 2)    # modulus 取模/求余
print(2 ** 3)   # exponent 幂运算
```

# 判断奇数偶数

```python
num = 15

# 判断奇数偶数 Odd / Even
if num % 2 == 0:
    print(num, "是偶数 Even")
else:
    print(num, "是奇数 Odd")

# 判断3的倍数 Divisible by 3
if num % 3 == 0:
    print(num, "是3的倍数 Divisible by 3")
else:
    print(num, "不是3的倍数 Not divisible by 3")
```

```python
# Why Learn Data Science?
Data is growing fast, and storage & computing costs are dropping.
Data science helps find insights and support smart decisions.
Data scientists are in high demand and well-paid.
It is widely used in many industries.

Data is everywhere, and data science helps us understand it.
It lets us find patterns, make predictions, and solve real problems.
It has high demand in many fields like business, medicine, and technology.
It provides good job opportunities and strong career growth.
It helps people and companies make smarter decisions.
```

```python
# What you can learn from Data Science
Python programming: Learn to code and process data with Python.
Statistics and math: Understand data rules and analyze numbers.
Data analysis and visualization: Find insights and show data with charts.
Machine learning basics: Build simple predictive models.
Communication and domain knowledge: Explain results and solve real problems
```

```python
# What Data Scientists Do
Collect, process and prepare raw data.
Ensure data quality and security.
Analyze and visualize data to discover insights.
Store and manage data for future use.
Work in teams to solve problems using data.
```

```python
# Key Differences Between Data Scientists & Data Engineers
1. Data scientists analyze data, while data engineers build data systems.
2. Data scientists create models, while data engineers build data pipelines.
3. Data scientists find insights, while data engineers ensure data quality.
4. Data scientists visualize data, while data engineers manage databases.
```

```python
# Key Term's Explanation
for loop: Repeats code for each item in a list or group.
while loop: Repeats code as long as a condition is true.
if statement: Runs different code based on true or false conditions.
function: A reusable block of code for a specific task.
list: An ordered collection of multiple items.
dictionary: Stores data as key-value pairs for organized information.
```

```python
# What is data science, and what disciplines does it integrate?
Data science is an interdisciplinary field that combines:
Math & Theory (statistics, linear algebra, optimization)
Applied Algorithms (machine learning, data structures)
Engineering & Technologies (storage/computing platforms, tools)
Domain Expertise (finance, text, images, econometrics)
Art (visualization, infographics)
Best Practices (data cleaning, missing value handling)
```

```python
# Why is data science a multi-disciplinary field, and why are all these pillars necessary?
Data science requires a full skill set:
Math/theory provides the foundational logic for analysis.
Algorithms turn theory into actionable solutions.
Engineering enables scalable big data processing.
Domain expertise ensures insights are business-relevant.
Visualization communicates results to stakeholders.
Best practices ensure data quality and reliable analysis.
No single pillar is sufficient; all are required for end-to-end data work.
```

```python
# What core skills does a data scientist need to master, and how do they work together?
A data scientist needs 6 core skills:
Strong mathematical/statistical foundation
Proficiency in algorithms and programming
Engineering skills for data infrastructure
Deep domain knowledge of the industry
Visualization and communication skills
Practical data handling experience
These skills work together to turn raw data into actionable, business-aligned insights.
```
