# Creating Series

```python
list=[10,20,30,40,50]
label=['a','b','c','d','e']
s1=pd.Series(list,index=label)
print(s1)
```

# function application and mapping

```python
# 可以直接用series计算，不用转为dataframe
series=pd.Series([1,2,3,4,5])
np.sqrt(series)
```

```python
dict={'A':[1,2,3],'B':[4,5,6]}
df=pd.DataFrame(dict)
np.square(df)
```

# apply

```python
series=pd.Series([1,2,3,4,5])
series.apply(np.sqrt)

# np.sqrt(series)
```

```python
def categorize(x):
    if x >= 5:
        return "high"
    else:
        return"low"
```

```python
# apply+function 为什么不能用于行？
df['B'].apply(categorize)
```

# map()

```python
dict={'A':[1,2,3],'B':[4,5,6]}
df=pd.DataFrame(dict)
df.map(np.square)
# np.square(df)  # 一样的结果
```

```python
df['B'].apply(categorize)
```

```python
df['B'].map(categorize)
```

```python
df=pd.DataFrame({'Month':['Jan','Feb','Mar','Apr'],
                  'Rev':[230,300,400,500]
})
df
```

```python
df['Month'].map({'Jan':1,'Mar':99,"Mar":2,"Apr":4})
```

```python
dict={'A':[1,2,3],'B':[4,5,6]}
df=pd.DataFrame(dict)
df.apply(sum,axis=1)  # 但是map不行，apply 和map的区别需要特别注意
```

# Pivot Tables and Cross Tabulations

# Pivot Tables

```python
data={
    'City':['KL','Selongor','Ipoh','Kuching','KL','KL'],
    'Category':['A','A','B','B','B','B'],
    'Sales':[2000,3000,4000,5000,3000,1000]
}

df=pd.DataFrame(data)
df
```

```python
# Default aggfunc=mean
pd.pivot_table(df,values='Sales',index='Category',columns='City')
```

```python
pd.pivot_table(df,values='Sales',index='Category',columns='City',aggfunc='sum')
```

# Cross Tabulations

```python
pd.crosstab(df['City'],df['Category'])
```

# View and Copy

```python
# Pandas 3.0 already remove the warning 
df=pd.DataFrame({'A':[1,2,3],'B':[4,5,6],'C':[7,8,9]})

# create a view
view=df[['A','B']]

# modify the view
view['A']=9999

print(df)
print("\n")
print(view)
```
