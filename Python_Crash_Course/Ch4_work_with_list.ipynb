# List + Statistics

```python
digits = [1, 2, 3, 4, 5, 6, 7, 8, 9, 0]
print(min(digits))
print(max(digits))
print(sum(digits))
print(len(digits))
print(sorted(digits)) # 顺序排列
print(list(reversed(digits))) # 倒序排列
```

# list + slicing

```python
# list 从0开始
players = ['charles', 'martina', 'michael', 'florence', 'eli']
print(players[1:3]) # 1-2 共2个(0,1)
print(players[:3]) # 0-2 共3个(0,1,2)
print(players[2:]) # 2-结尾 共3个(2,3,4)
print(players[-2:]) # -2-结尾 共2个(-2,-1)
```

# list + slicing + for loop

```python
players = ['charles', 'martina', 'michael', 'florence', 'eli'] 
print("Here are the first three players on my team:")
for player in players[:3]:
    print(player.title())
```

# Copying a List

```python
my_foods = ['pizza', 'falafel', 'carrot cake']
friend_food=my_foods[:] # 必须加冒号：

my_foods.append('ice cream')
friend_food.append('beef')
friend_food.remove('pizza')

print("My favorite foods:")
print(my_foods)

print("\nMy friend's favorite foods are:")
print(friend_food)
```

# Excercise

```python
# 4-10 list + slice
players = ['charles', 'martina', 'michael', 'florence', 'eli'] 

# 前三个
print("The first three items in the list are:")
print(players[:3])

# 中间三个
print("\nThree items from the middle of the list are:")
print(players[1:4])

# 最后三个
print("\nThe last three items in the list are:")
print(players[-3:])
```

```python
# 4-11 list + copy
my_pizzas=['beef','eggs','lamb','fish']
friend_pizzas=my_pizzas[:]

my_pizzas.append('mango')
friend_pizzas.append('orange')

print("\nMy favorite pizzas are:")
for my_pizza in my_pizzas:
    print(my_pizza.title())

print("\nMy friend’s favorite pizzas are:")
for friend_pizza in friend_pizzas:
    print(friend_pizza.title())
```

```python
# 4-12 多个list
my_pizzas=['beef','eggs','lamb','fish']
my_foods = ['pizza', 'falafel', 'carrot cake']

print("\nFirst list:")
for my_pizza in my_pizzas:
    print(my_pizza.title() + "taste very delicious.")

print("\nAnother list:")
for my_food in my_foods:
    print("I like eating " + my_food.title())
```

#  Tuple(元组)--圆括号/不能改值、只能重新定义

```python
dimensions = (200, 50)
print(dimensions[0])
print(dimensions[1])
```

```python
# 不能变更，只能重新定义
#原始tuple
dimensions1 = (200, 50)
print(dimensions1)

# 重新定义
dimensions2 = (450, 50)
print(dimensions2)
```
