```python
alien_0={"x_position":0,"y_position":10,"speed":"medium"}
print("Original x_position: "+str(alien_0['x_position']))

if alien_0["speed"]=='slow':
    x_increment = 1
elif alien_0["speed"]=='medium':
    x_increment = 2
else:
    x_increment = 3

alien_0["x_position"]=alien_0["x_position"]+x_increment
print("New x_position: " + str(alien_0["x_position"]))
```

```python
# 薄弱题
alien_0={"x_position":0,"y_position":10,"speed":"medium"}
alien_0["speed"]="fast"
print("Original x_position: "+str(alien_0['x_position']))

if alien_0["speed"]=='slow':
    x_increment = 1
elif alien_0["speed"]=='medium':
    x_increment = 2
else:
    x_increment = 3

alien_0["x_position"]=alien_0["x_position"]+x_increment
print("New x_position: " + str(alien_0["x_position"]))
```

# dict + range() + list--批量创建字典

```python
aliens=[]

for alien_num in range(30):
    new_alien={"color":"green","points":5,"speed":"slow"}
    aliens.append(new_alien)

for alien in aliens[:5]:
    print(alien)
print("...")

print("Total number of the aliens are: "+str(len(aliens)))
```

# # dict + range() + list + for--修改字典

```python
# Make an empty list for storing aliens.
aliens=[]

# Make 30 green aliens.
for alien in range(30):
    new_alien={'color': 'green', 'points': 5, 'speed': 'slow'}
    aliens.append(new_alien)

# Show the first 5 aliens:
for alien in aliens[0:5]:
    print(alien)
print("...")

# to change the first 3 aliens
for alien in aliens[0:3]:
    if alien['color']=='green':
        alien['color']='blue'
        alien['speed']='medium'
        alien['points']=10

# output the update aliens
print("Update aliens:")
for alien in aliens[0:5]:
    print(alien)
```

```python
# 薄弱题
# crate a empty list for storing
students=[]

# create 20 student info
for student in range(20):
    student_info={'name':'student','grade':80,'class':1}
    students.append(student_info)

# output first 5 students info
for student in students[0:5]:
    print(student)
print("...")

for student in students[0:3]:
    if student['grade']==80:
        student['grade']=95
        student['class']=2

print("student info update:")
for student in students[0:5]:
    print(student)
print("...")
```

```python
# create an empty list to store
items=[]

# create 25 items
for item in range(25):
    item_info={'type': 'potion', 'power': 10, 'price': 50}
    items.append(item_info)

# output first 5 items
for item in items[0:5]:
    print(item)
print("...") 

# update information
for item in items[0:2]:
    if item['type']=='potion':
        item['type']='sword'
        item['power']=50
        item['price']=200

# output the update itemds
for item in items[0:5]:
    print(item)
print("...") 
```

```python
# 6-11 dict in dict
cities={
    'beijing':{'Country':'China','Population':'5 million','Fact':'capital of China'},
    'kuala lumpur':{'Country':'Malaysia','Population':'2 million','Fact':'a hot weather city'},
    'newyork':{'Country':'America','Population':'3.5 million','Fact':'a developed city'},
}

for city_name in cities.keys():
    print("\n"+city_name.title()+" has these info:")
    for key,value in cities[city_name].items():
        print(" -"+key+": "+value.title())
```

```python
# 6-12 dict in dict + 新增 key-value pair
cities={
    'beijing':{'Country':'China','Population':'5 million','Fact':'capital of China'},
    'kuala lumpur':{'Country':'Malaysia','Population':'2 million','Fact':'a hot weather city'},
    'newyork':{'Country':'America','Population':'3.5 million','Fact':'a developed city'},
}

cities['beijing']['food'] = 'Peking Duck'
cities['kuala lumpur']['food'] = 'Nasi Lemak'
cities['newyork']['food'] = 'Pizza'

for city_name, info in cities.items():
    print("\n"+city_name.title()+" :")
    for key, value in info.items():
        print("-" + key + ":"+ value)
```
