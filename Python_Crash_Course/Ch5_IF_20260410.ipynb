# # if + epmty list

```python
usernames=[]  # 空list，不能写成usernames=[‘’]
if usernames:   # 如果usernames list 非空时执行 if语句，是空的话，执行else语句
    for username in usernames:
        if username=='admin':
            print("Hello admin, would you like to see a status report?")
        else:
            print("Hello " + username.title() +", thank you for log-ging again.")
else:
    print("We need to find some users!")
```

# if + Multiple Lists

```python
available_toppings = ['mushrooms', 'olives', 'green peppers','pepperoni', 'pineapple', 'extra cheese']
requested_toppings = ['mushrooms', 'french fries', 'extra cheese']

for requested_topping in requested_toppings:
    if requested_topping in available_toppings:
        print("Adding " + requested_topping +".")
    else:
        print("Sorry, we don't have " + requested_topping + ".")

print("\nFinished making your pizza!")
```

# 5-10 if + 多list + lower/upper() 

```python
# 5-10 薄弱题
current_users=['admin','mica','DONNA','maria','ANNA']
new_users=['lily','anna','donna','jack','MICA']

# 关键转换
current_users_lower=[user.lower() for user in current_users]

for new_user in new_users:
    if new_user.lower() in current_users_lower:
        print(new_user+" has already been used,you need to enter a new username!")
    else:
        print(new_user+" the username is available.")
```

# 5-11 if-elif-else + range()

```python
# 5-11
nums=list(range(1,11))
for num in nums:
    if num==1:
        print("1st")
    elif num==2:
        print("2nd")
    elif num==3:
        print("3rd")
    else:
        print(str(num)+"th")
```
