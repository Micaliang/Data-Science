# Introduction lists(从头是0，结尾是-1)

```python
bicycles = ['trek', 'cannondale', 'redline', 'specialized']
print(bicycles)
```

```python
print(bicycles[:]) # 输出整个列表
print(bicycles[1:3]) # python是从0开始，左闭右开(不含3)
print(bicycles[1:]) # 从1到结尾
print(bicycles[:3]) # 从 0-3(不含3)
print(bicycles[-1]) # 从右到左，右边最后一个数(尾数从-1开始)
print(bicycles[-2].title())
```

# sort()--按字母顺序输出

```python
# 正序(按字母顺序)
cars = ['bmw', 'audi', 'toyota', 'subaru']
cars.sort() # 永久
print(cars)
```

```python
# 逆序(按字母顺序)
cars = ['bmw', 'audi', 'toyota', 'subaru']
cars.sort(reverse=True)
print(cars)
```

# sorted()--不改变原列表顺序

```python
cars = ['bmw', 'audi', 'toyota', 'subaru']
print("Here is the original list:")
print(cars)

print("\nHere is the sorted list:")
print(sorted(cars)) # sorted(cars)临时改变顺序，如果用cars.sort()则是永久改变顺序

print("\nHere is the original list again:")
print(cars)
```

```python
cars = ['bmw', 'audi', 'toyota', 'subaru']

# 按字母倒序输出
result = sorted(cars, reverse=True)
print(result)
```

# reverse()--逆序输出(不管字母顺序)

```python
cars = ['bmw', 'audi', 'yamha', 'subaru']

# 原始顺序输出
print(cars)

# 逆序输出
cars.reverse()
print(cars)

# 再逆序一次就是原始顺序输出
cars.reverse()
print(cars)
```

# Excercise

```python
# 3-8
places=['kuala lumpur','new york','london','Beijing','shanghai']

# 原顺序输出
print("Original order lists:")
print(places)

# 按字母正序输出(不改变原列表)
print("\nSorted order lists:")
print(sorted(places))

# 原顺序输出
print("\nOriginal order lists again:")
print(places)

# 按字母逆序输出(不改变原列表)
print("\nReverse alphabetical order lists:")
print(sorted(places,reverse=True))

# 原顺序输出
print("\nOriginal order lists:")
print(places)

# 列表反转（颠倒原有顺序，不按字母）
print("\nUse reverse() to change the lists:")
places.reverse()
print(places)

# 再次反转，恢复原来顺序
print("\nBack to its original order:")
places.reverse()
print(places)

# 按字母正序排列（永久修改原列表）
print("\nStored in alphabetical order:")
places.sort()
print(places)

# 按字母逆序排列（永久修改原列表）
print("\nStored in reverse alphabetical order:")
places.sort(reverse=True)
print(places)
```

```python
# 3-10 Every Function
# 创建一个列表（我用城市列表）
cities = ['tokyo', 'paris', 'london', 'sydney', 'dubai']

# 1. 打印原始列表
print("Original list:")
print(cities)

# 2. 访问列表元素（索引）
print("\nFirst city:", cities[0].title())

# 3. 修改元素
cities[1] = 'rome'
print("\nAfter changing city:", cities)

# 4. 添加元素到末尾 append()
cities.append('berlin')
print("\nAfter append:", cities)

# 5. 插入元素到指定位置 insert()
cities.insert(2, 'madrid')
print("\nAfter insert:", cities)

# 6. 删除元素 del
del cities[0]
print("\nAfter del:", cities)

# 7. 弹出末尾元素 pop()
popped_city = cities.pop()
print("\nPopped city:", popped_city)
print("Now list:", cities)

# 8. 根据值删除 remove()
cities.remove('madrid')
print("\nAfter remove:", cities)

# 9. 临时排序 sorted()
print("\nSorted list:", sorted(cities))

# 10. 永久排序 sort()
cities.sort()
print("\nAfter sort():", cities)

# 11. 反向永久排序 sort(reverse=True)
cities.sort(reverse=True)
print("\nAfter sort(reverse=True):", cities)

# 12. 反转列表 reverse()
cities.reverse()
print("\nAfter reverse():", cities)

# 13. 获取列表长度 len()
print("\nTotal number of cities:", len(cities))
```
