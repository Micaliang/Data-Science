```python
# 遍历 1 到 10，计算每个数的平方，存入列表 squares 并打印
squares=[]
for num in nums:
    squares.append(num**2)
print(squares)
```

# 三、if 条件判断

```python
# 输入学生分数（0–100）
# 90 及以上 → 输出 A
# 80–89 → B
# 70–79 → C
# 60–69 → D
# 60 以下 → F
```

```python
mark=float(input("please input student's marks:"))
if mark >= 90:
    print("A")
elif 80<= mark <=89:
    print("B")
elif 70<= mark <=79:
    print("C")
elif 60<= mark <=69:
    print("D")
else:
    print("F")
```

```python
# 如果分数 >100 或 <0，输出 Invalid score
# 否则按等级输出：
# ≥85 → A
# ≥75 → B
# ≥60 → C
# <60 → F
```

```python
mark=float(input("please input student's marks:"))
if mark<0 or mark>100:
    print("Invalid score")
elif mark<60:
    print("F")
elif 60<=mark<75:
    print("C")
elif 75<=mark<85:
    print("B")
else:
    print("A")
```

# 四、字典（Dictionary）

```python
# 创建字典 student，包含键值对："name":"Tom"、"age":20、"grade":"A"
student={"name":"Tom","age":20,"grade":"A"}
print(student)
```

```python
# 打印 student 中 "name" 对应的值
print(student["name"])
```

```python
# 添加键值对 "city":"KL" 到 student
student["city"]="KL"
print(student)
```

```python
# 修改 student 中 "age" 为 21
student["age"]=21
print(student)
```

```python
# 删除 student 中 "grade" 键值对
del student["grade"]
print(student)
```

```python
# 遍历 student 所有键 + 值并打印
for key,value in student.items():
    print("\nKey:"+key)
    print("Value:"+str(value))
```

```python
for k,v in student.items():
    print(k,v)
```

# 六、while 循环

```python
# 用 while 循环打印 1 到 5
n=1
while n<6:
    print(n)
    n+=1
```

```python
# 循环到 3 时用 break 停止
n=1
while n<6:
    if n==3:
        break
    print(n)
    n+=1
```

```python
# 循环跳过 3，打印 1、2、4、5
n=0
while n<5:
    n+=1
    if n==3:
        continue
    print(n)
```

```python
# 用 flag 标志位控制循环，从 10 倒数到 1 停止
n=10
flag=True
while flag:
    print(n)
    if n==1:
        flag=False
    n-=1
```
