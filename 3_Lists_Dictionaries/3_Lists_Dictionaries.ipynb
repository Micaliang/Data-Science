# while+continue

```python
# 标准模版
i = 0  # 初始值
while i < 上限:
    i += 1  # 先+1，非常重要！
    if i == 要跳过的数:
        continue  # 跳过，不打印
    print(i)
```

```python
i=0
while i<6:
    i+=1
    if i==3:
        continue
    print(i)
```

```python
prompt = "\nPlease enter the name of a city you have visited:"
prompt += "\n(Enter 'quit' when you are finished.) "

while True:
    city = input(prompt)
    if city == 'quit':
        break
    else:
        print("I'd love to go to " + city.title() + "!")
```

# while + flag

```python
# 标准模版
flag = True  # 开关打开
i = 开始的数

while flag:
    print(i)
    if i == 1:  # 到1就关掉开关
        flag = False
    i -= 1  # 倒数，每次减1
```

```python
prompt = "\nPlease enter the name of a city you have visited:"
prompt += "\n(Enter 'quit' when you are finished.) "

flag=True
while flag:
    city = input(prompt)
    if city == 'quit':
        flag=False
    else:
        print("I'd love to go to " + city.title() + "!")
```
